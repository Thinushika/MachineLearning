```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\TJ\Documents\Machine Learning\Data sets\weight and height\weight-height.csv')
```

```python
print(data.shape)
data.head()
```

```python
plt.style.use('ggplot')
data.Height.plot(kind='hist', color='green',edgecolor='black')
plt.title('Distribution of Height', size=24)
plt.xlabel('Height (inches)', size=18)
plt.ylabel('Frequency', size=18)
```

```python
data.Weight.plot(kind='hist', color='purple', edgecolor='black')
plt.title('Distribution of Weight', size=24)
plt.xlabel('Weight (pounds)', size=18)
plt.ylabel('Frequency', size=18);
```

```python
data.describe()
```

```python
ax1 = data[data['Gender'] == 'Male'].plot(kind='scatter', 
                                      x='Height', 
                                      y='Weight', 
                                      color='green', 
                                      alpha=0.5, 
                                      figsize=(10, 7))
data[data['Gender'] == 'Female'].plot(kind='scatter', 
                                  x='Height', 
                                  y='Weight', 
                                  color='red', 
                                  alpha=0.5, 
                                  figsize=(10 ,7), 
                                  ax=ax1)

plt.legend(labels=['Males', 'Females'])
plt.title('Relationship between Height and Weight', size=24)
plt.xlabel('Height (inches)', size=18)
plt.ylabel('Weight (pounds)', size=18);
```
